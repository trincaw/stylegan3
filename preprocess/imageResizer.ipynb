```python
import os
from PIL import Image
import cv2

# Set the paths for the input and output directories
input_dir = "C:\\Users\\nico\\Desktop\\preprocess\\input_images"
output_dir = "C:\\Users\\nico\\Desktop\\preprocess\\output_images"

def resize_image(input_dir, output_dir):
    # Ensure that the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Iterate over the files in the input directory
    for filename in os.listdir(input_dir):
        # Check if the image is already resized
        if filename.endswith(".png"):
            output_path = os.path.join(output_dir, filename)
            if os.path.exists(output_path):
                print(f"Image {filename} already resized. Skipping...")
                continue

        # Load the image using OpenCV
        image_path = os.path.join(input_dir, filename)
        print(image_path)
        image = cv2.imread(image_path)

        # Resize the image to 1024x1024
        resized_image = cv2.resize(image, (1024, 1024))

        # Convert the image to PIL format
        pil_image = Image.fromarray(resized_image)

        # Change the format to PNG and save with compression level
        output_filename = os.path.splitext(filename)[0] + ".png"
        output_path = os.path.join(output_dir, output_filename)
        pil_image.save(output_path, format='PNG')

resize_image(input_dir, output_dir)
```
